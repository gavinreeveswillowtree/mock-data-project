```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
from faker import Faker
import random
```

```python
fake = Faker()
fake_data = []

for i in range(1000):
    first_name = fake.first_name()
    last_name = fake.last_name()
    email = fake.email()
    area_code = fake.random_int(min=100, max=999, step=1) # random 3 digit number
    state = fake.state()
    country = fake.country()
    

    fake_data.append([first_name, last_name, email, area_code, state, country])

columns = ['First Name', 'Last Name', 'Email', 'Area Code', 'State', 'Country']
```

```python
df = pd.DataFrame(fake_data, columns=columns)

df 
```
